```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(0.50, 1.0, (250,2))
```

```python
X_test2 = np.random.rand(0.20, 0.90, (100, 2))
```

```python
X_test = np.random.randint(0, 1.0, (1000, 2))
print(X_test)
```

```python
plt.scatter(X[:, 0], X[:, 1])
# plt.scatter(range(len(X_test)), X_test, c='red')
# plt.scatter(range(100, len(X_test2)+100), X_test2, c='yellow')
plt.show()
```

```python
lof = LocalOutlierFactor(n_neighbors=2  ,novelty=True)
lof.fit(X)
```

```python
y_pred = lof.predict(X_test)
```

```python
y_pred.sum()
```

```python
y_pred3 = lof.predict(X_test2)
```

```python
y_pred3.sum()
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred)
# plt.scatter(range(120, 220), X_test2, c=y_pred3)
# plt.show()
```
